Below is an example in Python using TensorFlow and Keras to create a convolutional neural network (CNN) that classifies images of digits 0 to 9 using the MNIST dataset.

```python
#Paso 1: Importar las Librerías Necesarias
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import random
import numpy as np
# Configurar semillas aleatorias para reproducibilidad
seed_value = 1234
# 1. Configurar la semilla aleatoria de Python
random.seed(seed_value)
# 2. Configurar la semilla aleatoria de numpy
np.random.seed(seed_value)
# 3. Configurar la semilla aleatoria de TensorFlow
tf.random.set_seed(seed_value)
```

```python
#Paso 2: Cargar y Preprocesar el Conjunto de Datos MNIST
#MNIST es un conjunto de datos de 60,000 imágenes de entrenamiento y 10,000 imágenes de prueba de dígitos escritos a mano (0-9).
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
# Redimensionar las imágenes para que tengan un solo canal (escala de grises)
train_images = train_images.reshape((train_images.shape[0], 28, 28, 1))
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1))
# Normalizar los valores de píxeles entre 0 y 1
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255
```

```python
#Paso 3: Construir el Modelo de la CNN
from tensorflow.keras import regularizers
# Construir el modelo de la CNN
model = models.Sequential()

# Primera capa convolucional
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1),kernel_regularizer=regularizers.l2(0.001)))
model.add(layers.MaxPooling2D((2, 2)))

# Segunda capa convolucional
model.add(layers.Conv2D(64, (3, 3), activation='relu',kernel_regularizer=regularizers.l2(0.001)))
model.add(layers.MaxPooling2D((2, 2)))

# Tercera capa convolucional (opcional)
model.add(layers.Conv2D(64, (3, 3), activation='relu',kernel_regularizer=regularizers.l2(0.001)))

# Aplanar la salida para conectarla a la capa completamente conectada
model.add(layers.Flatten())
# Capa completamente conectada
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.15))
# Capa de salida con 10 neuronas (una por cada clase de dígito)
model.add(layers.Dense(10, activation='softmax'))
```

```python
#Paso 4: Compilar el Modelo
# Compilar el modelo
model.compile(optimizer='Adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Paso 5: Entrenar el Modelo
#from tensorflow.keras.callbacks import EarlyStopping
# Definir el EarlyStopping
#early_stopper = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(train_images, train_labels, epochs=5,
                    validation_data=(test_images, test_labels))
```

```python
# Guardar el modelo en formato SavedModel
model.save('./data/mymodelcnn.keras')
```

```python
#Paso 6: Evaluar el Modelo
# Evaluar el modelo en el conjunto de prueba
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"Precisión en el conjunto de prueba: {test_acc:.4f}")
```

```python
#Paso 7: Visualizar el Rendimiento
# Graficar la precisión y pérdida durante el entrenamiento
plt.plot(history.history['accuracy'], label='Precisión en entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión en validación')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Pérdida en entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida en validación')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()
plt.show()
```

```python
#Paso 8: Predecir un nuevo valor
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Paso 1Cargar el modelo completo
model_load = load_model('./data/mymodelcnn.keras')

# Paso 2: Cargar y preprocesar la imagen nueva
img_path = './images/minumero3.png'  # Reemplaza con la ruta a la imagen que quieres validar
img = image.load_img(img_path, target_size=(28, 28),color_mode='grayscale', interpolation='bicubic')  # Cargar la imagen y redimensionarla
img_array = image.img_to_array(img)  # Convertir la imagen a un array numpy
img_array = img_array / 255.0  # Escalar los valores de los píxeles entre 0 y 1
img_array = np.expand_dims(img_array, axis=0)  # Añadir una dimensión extra para el tamaño del lote
# Paso 3: Realizar la predicción
predicciones = model_load.predict(img_array)
print("Valores de prediccion")
print(predicciones)

# Paso 4: Interpretar la predicción
# Asumiendo que estás clasificando los dígitos del 0 al 9
#np.argmax: Esta función de NumPy devuelve el índice de la posición del valor máximo en un array.
#Si el array es multidimensional, por defecto, np.argmax aplanará el array en una sola dimensión y devolverá el índice del valor máximo.
clase_predicha = np.argmax(predicciones)
print(f'El modelo predice que la imagen es un {clase_predicha}.')

# Mostrar la imagen
plt.imshow(img, cmap='gray')
plt.title(f'Predicción: {clase_predicha}')
plt.show()
```
