### Preprocesar Datos- Manejo de datos perdidos


```python
# 1. descartar las columnas con demasiados valores pedridos
# STDs: Time since first diagnosis y STDs: Time since last diagnosi  >91 % de valores perdidos
# Eliminar columnas
columnas_a_eliminar = ['STDs: Time since first diagnosis', 'STDs: Time since last diagnosis']
data = data.drop(columnas_a_eliminar, axis=1)
```

```python
# Mostrar el DataFrame modificado
data.head()
```

```python
#2. Imputación de los valores perdidos
# Imputar los valores perdidos con la media de cada columna
data_filled = data.fillna(data.mean())
print(len(data_filled))
```

```python
# 3. datos duplicados en el dataset
#data_filled.drop_duplicates(inplace=True)
#print(len(data_filled))
```

```python
# Validar que no tengamos datos peerdidos
datos_perdidos = data_filled.isnull().sum()
print(datos_perdidos)
```

### Crear Modelo Base - RandomForest

```python
# Dividir los datos en características (X) y etiquetas (y)
x= data_filled.drop('Dx:Cancer', axis=1)
y = data_filled['Dx:Cancer']
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
## Optimizacion de parametros
# Instancia del modelo
rf_model = RandomForestClassifier(random_state=42)

# Definir los hiperparámetros y sus posibles valores
param_grid = {
    'n_estimators': [10,25,50],
    'max_depth' : [5,10,15],
    'criterion' : ['mse', 'mae', 'gini', 'entropy', 'log_loss'],
    'min_samples_split': [2,4,6],
    'min_samples_leaf': [1,2,4],
}
# Crear el objeto GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='accuracy')
# Ajustar el modelo con GridSearchCV
grid_search.fit(X_train, y_train)

# Obtener el modelo con el mejor rendimiento
best_model = grid_search.best_estimator_
```

```python
# Mejores parametros del modelo
grid_search.best_params_
```

```python
y_train_pred= best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)
accuracy_train = accuracy_score(y_train,y_train_pred)
accuracy_test = accuracy_score(y_test,y_test_pred)
print("Train:",accuracy_train)
print("Test",accuracy_test)
```

```python
# Mostrar el informe de clasificación wn test
print("Clasificacion en test")
conf_matrix = confusion_matrix(y_test, y_test_pred)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
plt.show()
print(classification_report(y_test, y_test_pred))

```

```python
# Mostrar el informe de clasificación wn trainig
print("Clasificacion en training")
conf_matrix1 = confusion_matrix(y_train, y_train_pred)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix1, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
plt.show()
print(classification_report(y_train, y_train_pred))
```

Precisión (Precision): Representa la proporción de verdaderos positivos (TP) sobre el total de predicciones positivas (TP + FP). Es una medida de la precisión del modelo en la predicción de muestras positivas.

Recuperación (Recall): Representa la proporción de verdaderos positivos (TP) sobre el total de muestras positivas reales (TP + FN). Es una medida de la capacidad del modelo para encontrar correctamente las muestras positivas.

Puntuación F1 (F1-Score): Es una medida que combina tanto la precisión como la recuperación en un solo valor. El F1-Score es la media armónica de precisión y recuperación y proporciona una medida equilibrada del rendimiento del modelo.

Soporte (Support): Representa el número de muestras reales en cada clase. Proporciona información sobre el desequilibrio de clases y la cantidad de datos utilizados para calcular las métricas.
