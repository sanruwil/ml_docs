## Crete Base Model  
####  1. RandomForest

```python
# Split the data into features (X) and labels (y)  
x= data_sin_outliers.drop('Dx:Cancer', axis=1)
y= data_sin_outliers['Dx:Cancer']
# Split the data into training(70%) and test(30%) sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
def buil_model(model,randomsampler=None):
    # Training the model with sampler
    if randomsampler is not None:
        X_train_resampled, y_train_resampled = randomsampler.fit_resample(X_train, y_train)
        model.fit(X_train_resampled, y_train_resampled)
    else:   
        model.fit(X_train, y_train) # Training model without sampler
    # Forecasting data test ( Y)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    # Calculate accuracy in train data
    train_acc = accuracy_score(y_train,train_pred)
    # Calculate accuracy in test data
    test_acc  = accuracy_score(y_test,test_pred)
    print("The accuracy  Train is:",train_acc)
    print("The accuracy  Test es:",test_acc)
    return train_pred,test_pred
```

```python
# Create the model Random Forest
rf_model = RandomForestClassifier(n_estimators=50, random_state=42)
rf_train_pred,rf_test_pred= buil_model(rf_model)
```

#### Feature importance analysis

```python
feature_scores = pd.DataFrame(pd.Series(rf_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)).T
plt.figure(figsize=(12,6))
sns.barplot(data=feature_scores)
i=0
for index, value in enumerate(feature_scores.values.flatten()):
    if(i<4):
      plt.annotate(f'{value:.2f}', xy=(index, value), va='bottom')
    i=i+1
plt.title("Factores clave en la predicción de Cancer")
plt.xticks( rotation='vertical')
plt.show()
print("Features more important in RandomForest")
pd.DataFrame(feature_scores).T
```

```python
def plot_metrics(y_test,test_pred, model):
    # Show report clasification test 
    print("Report clasification  TEST --->"+model)
    conf_matrix = confusion_matrix(y_test, test_pred)
    # Create the heatmap diagraman using  seaborn
    sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", cbar=False)
    # setup labels of the axis  x e y
    plt.xlabel('Label Forecast')
    plt.ylabel('Label Real')
    myreport=classification_report(y_test, test_pred)
    print(myreport)
    plt.show()
    return myreport
```

```python
rf_base_report = plot_metrics(y_test, rf_test_pred,"RandomForest")
```

Precision: Represents the proportion of true positives (TP) over the total of positive predictions (TP + FP). It is a measure of the model's accuracy in predicting positive samples.(Prediction Columns)

Recall: Represents the proportion of true positives (TP) over the total of actual positive samples (TP + FN). It is a measure of the model's ability to correctly identify positive samples. (Actual Rows)

F1-Score: It is a measure that combines both precision and recall into a single value. The F1-Score is the harmonic mean of precision and recall, providing a balanced measure of the model's performance.

Support: Represents the number of real samples in each class. It provides information about class imbalance and the amount of data used to calculate the metrics.

#### 2.SVM(Suppport Vector Machine)

```python
from sklearn.svm import SVC
#Create the model of  SVM
svm_model = SVC(kernel='linear')
svm_train_pred,svm_test_pred= buil_model(svm_model)
# Performance model
svm_base_report = plot_metrics(y_test, svm_test_pred,"SVM")
```

#### 3. LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
# Craete and Traind Model with  LogisticRegression
lr_model = LogisticRegression(penalty="l1",
    solver="liblinear",tol=1e-6,max_iter=int(1e6),warm_start=True,
    intercept_scaling=10000.0)
lr_train_pred,lr_test_pred= buil_model(svm_model)
# Performance model
lr_base_report = plot_metrics(y_test, lr_test_pred,"LogisRegresion")
```

#### 4. Neural Network

```python
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import StandardScaler
# Normalize data
scaler = StandardScaler()
X_trainN = scaler.fit_transform(X_train)
X_testN = scaler.transform(X_test)
# Define the architecture of the neural network
nn_model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')])
# Compile model
nn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
# Training model
nn_model.fit(X_trainN, y_train, epochs=100, batch_size=32, verbose=1)
# Evaluate the model on the test set
loss, accuracy = nn_model.evaluate(X_testN, y_test, verbose=0)
# forescating test
nn_test_pred = nn_model.predict(X_testN).flatten()
# Convert the predictions into a binary scale.
#In this case, any value above 0.5 is rounded to 1, and any value below 0.5 is rounded to 0.
nn_test_pred_binary = np.round(nn_test_pred)
```

```python
print("Test set loss -  Accuracy: ", loss, accuracy)
# Performance model
nn_base_report = plot_metrics(y_test, nn_test_pred_binary,"NN")
```

#### 5. K-Nearest Neighbors
K-Nearest Neighbors is a supervised learning algorithm. Where the data is 'trained' with data points corresponding to their classification. To predict the class of a given data point, it takes into account the classes of the 'K' nearest data points and chooses the class in which the majority of the 'K' nearest data points belong to as the predicted class.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

# Normalize data
scaler = StandardScaler()
X_traink = scaler.fit_transform(X_train)
X_testk = scaler.transform(X_test)
# number of nearest data points
k = 5
#Create model
knn_model = KNeighborsClassifier(n_neighbors = k)
# Training model
knn_model.fit(X_traink,y_train)
# forescating test
knn_test_pred = knn_model.predict(X_testk)
print("Train set Accuracy: ", metrics.accuracy_score(y_train, knn_model.predict(X_traink)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, knn_test_pred))
# Performance model
knn_base_report = plot_metrics(y_test, knn_test_pred,"K-Nearest Neighbors")
```

#### Plot  ROC Curves  Models 

```python
from sklearn.metrics import roc_curve, auc
# Calcular la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR)
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_test_pred)
svm_fpr, svm_tpr, _ = roc_curve(y_test, svm_test_pred)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_test_pred)
nn_fpr, nn_tpr, _ = roc_curve(y_test, nn_test_pred_binary)
knn_fpr, knn_tpr, _ = roc_curve(y_test, knn_test_pred)

# Calcular el área bajo la curva ROC (AUC)
roc_auc_rf = auc(rf_fpr, rf_tpr)
roc_auc_svm = auc(svm_fpr, svm_tpr)
roc_auc_lr = auc(lr_fpr, lr_tpr)
roc_auc_nn = auc(nn_fpr, nn_tpr)
roc_auc_knn = auc(knn_fpr, knn_tpr)

# Graficar la curva ROC
fig, ax = plt.subplots()
#ax.set_aspect('equal', adjustable='box')
line, = ax.plot(rf_fpr, rf_tpr, label='RandomForest (AUC = %0.2f)' % roc_auc_rf)
line, = ax.plot(svm_fpr, svm_tpr, label='SVN (AUC = %0.2f)' % roc_auc_svm)
line, = ax.plot(lr_fpr, lr_tpr, label='LogisticRegression (AUC = %0.2f)' % roc_auc_lr)
line, = ax.plot(nn_fpr, nn_tpr, label='NN (AUC = %0.2f)' % roc_auc_nn)
line, = ax.plot(knn_fpr, knn_tpr, label='K-Nearest Neighbors (AUC = %0.2f)' % roc_auc_knn)

plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc='lower right')

plt.show()
```

### Improving Random Forest Model - Handling Class Imbalance
#### Option 1. OverSampling
In this example, the imbalanced-learn (imblearn) library is used, which provides implementations of resampling techniques to address class imbalance. Specifically, SMOTE is used to generate synthetic examples of the minority class and balance the classes in the training set.
https://www.aprendemachinelearning.com/clasificacion-con-datos-desbalanceados/

```python
# Realizar sobremuestreo con SMOTE
oversampling = RandomOverSampler(random_state=42)
rf_model1 = RandomForestClassifier(n_estimators=50,random_state=42)
rf_train_pred1,rf_test_pred1= buil_model(rf_model1,oversampling)
# Generar informe de clasificación
rf1_base_report =plot_metrics(y_test, rf_test_pred1,"RandomForest-Oversampling")
```


#### Option 2. UnderSampling- Reduce the majority class


```python
# Realizar sobremuestreo con SMOTE
undersample = RandomUnderSampler(random_state=42)
rf_model2 = RandomForestClassifier(n_estimators=50,random_state=42)
rf_train_pred2,rf_test_pred2= buil_model(rf_model2,undersample)
# Generar informe de clasificación
rf2_base_report=plot_metrics(y_test, rf_test_pred2,"RandomForest-UnderSampling")
```


####  Option 3. One algorithm undersampling an another oversampling over  dataset

```python
smoteenn = SMOTETomek(random_state=42)
rf_model3 = RandomForestClassifier(n_estimators=50,random_state=42)
rf_train_pred3,rf_test_pred3= buil_model(rf_model3,smoteenn)
# Generar informe de clasificación
rf3_base_report=plot_metrics(y_test, rf_test_pred3,"RandomForest-JOIN(OverSampling-UnderSampling)")
```
