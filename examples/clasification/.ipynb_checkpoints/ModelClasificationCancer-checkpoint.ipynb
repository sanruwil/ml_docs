### Preprocesar Datos- Manejo de datos perdidos

Puedes utilizar métodos estadísticos como el rango intercuartílico (IQR) o la desviación estándar para detectar valores atípicos. También puedes visualizar tus datos mediante gráficos de caja (box plots) o diagramas de dispersión para identificar patrones inusuales.

## Crear Modelo Base 
1. RandomForest

```python
# Dividir los datos en características (X) y etiquetas (y)   data_filled
x= data_sin_outliers.drop('Dx:Cancer', axis=1)
y = data_sin_outliers['Dx:Cancer']
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Crear el modelo de Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Entrenar el modelo
rf_model.fit(X_train, y_train)
```

```python
# Predecimos Y
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)

# Calculo el accuracy en train 
train_acc = accuracy_score(y_true=y_train,y_pred=y_train_pred)
# Calculo el accuracy en test 
test_acc  = accuracy_score(y_true=y_test,y_pred=y_test_pred)
print("El accuracy en train es:",train_acc)
print("El accuracy en test es:",test_acc)
```

```python
# Mostrar el informe de clasificación en test
print("Informe clasificacion en TEST")
conf_matrix = confusion_matrix(y_test, y_test_pred)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
print(classification_report(y_test, y_test_pred))
plt.show()

```

```python
# Mostrar el informe de clasificación wn trainig
print("Informe clasificacion en TRAINING")
conf_matrix1 = confusion_matrix(y_train, y_train_pred)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix1, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
print(classification_report(y_train, y_train_pred))
plt.show()
```

Precisión (Precision): Representa la proporción de verdaderos positivos (TP) sobre el total de predicciones positivas (TP + FP). Es una medida de la precisión del modelo en la predicción de muestras positivas.

Recuperación (Recall): Representa la proporción de verdaderos positivos (TP) sobre el total de muestras positivas reales (TP + FN). Es una medida de la capacidad del modelo para encontrar correctamente las muestras positivas.

Puntuación F1 (F1-Score): Es una medida que combina tanto la precisión como la recuperación en un solo valor. El F1-Score es la media armónica de precisión y recuperación y proporciona una medida equilibrada del rendimiento del modelo.

Soporte (Support): Representa el número de muestras reales en cada clase. Proporciona información sobre el desequilibrio de clases y la cantidad de datos utilizados para calcular las métricas.

#### 2.SVN(Suppport Vector Machine)

```python
from sklearn.svm import SVC

# Crear y entrenar el modelo SVM
svm_rf = SVC(kernel='linear')
svm_rf.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_train_predsvm = svm_rf.predict(X_train)
y_test_predsvm = svm_rf.predict(X_test)
# Calculo el accuracy en train 
svn_train_acc = accuracy_score(y_true=y_train,y_pred=y_train_predsvm)
# Calculo el accuracy en test 
svn_test_acc  = accuracy_score(y_true=y_test,y_pred=y_test_predsvm)
print("El accuracy en train es:",svn_train_acc)
print("El accuracy en test es:",svn_test_acc)

# Evaluar el rendimiento del modelo
report = classification_report(y_test, y_test_predsvm)
print("Classification Report Test:\n", report)
```

```python
from sklearn.metrics import roc_curve, auc
# Calcular la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR)
fpr, tpr, _ = roc_curve(y_test, y_test_pred)
fpr_svn, tpr_svn, _ = roc_curve(y_test, y_test_predsvm)

# Calcular el área bajo la curva ROC (AUC)
roc_auc = auc(fpr, tpr)
roc_auc_svn = auc(fpr_svn, tpr_svn)

# Graficar la curva ROC
plt.figure()
plt.plot(fpr, tpr, label='RandomForest (AUC = %0.2f)' % roc_auc)
plt.plot(fpr, tpr_svn, label='SVN (AUC = %0.2f)' % roc_auc_svn)
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

### Mejora  Modelo Random Forest - Manejando desbalanceo de clases
#### Opcion 1. OverSampling
En este ejemplo, se utiliza la biblioteca imbalanced-learn (imblearn) que proporciona implementaciones de técnicas de remuestreo para abordar el desbalanceo de clases. En particular, se utiliza SMOTE para generar ejemplos sintéticos de la clase minoritaria y equilibrar las clases en el conjunto de entrenamiento.
https://www.aprendemachinelearning.com/clasificacion-con-datos-desbalanceados/

```python
# Cargar datos preprocesados
# data_filled

# Dividir los datos en conjuntos de entrenamiento y prueba
#X_train, X_test, y_train, y_test 

# Realizar sobremuestreo con SMOTE
smote = RandomOverSampler(random_state=42)
X_train_resampled1, y_train_resampled1 = smote.fit_resample(X_train, y_train)

# Entrenar un clasificador de RandomForest con los datos sobremuestreados
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_resampled1, y_train_resampled1)

# Predecimos Y
y_train_pred1 = clf.predict(X_train)
y_test_pred1 = clf.predict(X_test)

# Calculo el accuract en train 
train_acc1 = accuracy_score(y_true=y_train,y_pred=y_train_pred1)
# Calculo el accuract en test 
test_acc1  = accuracy_score(y_true=y_test,y_pred=y_test_pred1)
print("El accuracy en train es:",train_acc1)
print("El accuracy en test es:",test_acc1)

# Generar informe de clasificación
conf_matrix2 = confusion_matrix(y_test, y_test_pred1)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix2, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
print("Reporte en test:")
print(classification_report(y_test, y_test_pred1))
plt.show()

```

### Mejora  Modelo Random Forest - Manejando desbalanceo de clases
#### Opcion 2. UnderSampling- Reducir la clase mayoritaria


```python
# Cargar datos preprocesados
#data_filled
# Dividir los datos en conjuntos de entrenamiento y prueba
#X_train, X_test, y_train, y_test 
# Realizar submuestreo
undersample = RandomUnderSampler(random_state=42)
X_train_resampled2, y_train_resampled2 =  undersample.fit_resample(X_train, y_train)

# Entrenar un clasificador de RandomForest con los datos sobremuestreados
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_resampled2, y_train_resampled2)

# Predecimos Y
y_train_pred2 = clf.predict(X_train)
y_test_pred2 = clf.predict(X_test)

# Calculo el accuract en train 
train_acc2 = accuracy_score(y_true=y_train,y_pred=y_train_pred2)
# Calculo el accuract en test 
test_acc2  = accuracy_score(y_true=y_test,y_pred=y_test_pred2)
print("El accuracy en train es:",train_acc2)
print("El accuracy en test es:",test_acc2)

# Generar informe de clasificación
conf_matrix3 = confusion_matrix(y_test, y_test_pred2)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix3, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
print("Reporte en test:")
print(classification_report(y_test, y_test_pred2))
plt.show()
```

### Mejora  Modelo Random Forest - Manejando desbalanceo de clases
####  Opcion 3. un algoritmo de subsampling y otro de oversampling a la vez al dataset

```python
# Cargar datos preprocesados
#data_filled

# Separar características (X) y etiquetas (y)
# Dividir los datos en conjuntos de entrenamiento y prueba
#X_train, X_test, y_train, y_test
# Realizar submuestreo

smoteenn = SMOTETomek(random_state=42)
X_train_resampled3, y_train_resampled3 =  smoteenn.fit_resample(X_train, y_train)
# Entrenar un clasificador de RandomForest con los datos sobremuestreados
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_resampled3, y_train_resampled3)

# Predecimos Y (training y test)
y_train_pred3 = clf.predict(X_train)
y_test_pred3 = clf.predict(X_test)

# Calculo el accuract en train 
train_acc3 = accuracy_score(y_true=y_train,y_pred=y_train_pred3)
# Calculo el accuract en test 
test_acc3  = accuracy_score(y_true=y_test,y_pred=y_test_pred3)
print("El accuracy en train es:",train_acc3)
print("El accuracy en test es:",test_acc3)

# Generar informe de clasificación
conf_matrix4 = confusion_matrix(y_test, y_test_pred3)
# Crear el diagrama de calor usando seaborn
sns.heatmap(conf_matrix4, annot=True, cmap="Blues", fmt="d", cbar=False)
# Configurar etiquetas de los ejes x e y
plt.xlabel('Etiquetas Predichas')
plt.ylabel('Etiquetas Verdaderas')
# Mostrar el diagrama de calor
print("Reporte en test:")
print(classification_report(y_test, y_test_pred3))
plt.show()

```

#### Graficar la curva ROC

```python
from sklearn.metrics import roc_curve, auc
# Calcular la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR)
fpr_0, tpr_0, _ = roc_curve(y_test, y_test_pred)
fpr_1, tpr_1, _ = roc_curve(y_test, y_test_pred1)
fpr, tpr, _ = roc_curve(y_test, y_test_pred3)

# Calcular el área bajo la curva ROC (AUC)
roc_auc = auc(fpr, tpr)
roc_auc_o = auc(fpr_0, tpr_0)
roc_auc_1 = auc(fpr_1, tpr_1)
# Graficar la curva ROC
plt.figure()
plt.plot(fpr, tpr, label='Random-SMOTETomek (AUC = %0.2f)' % roc_auc)
plt.plot(fpr_1, tpr_1, label='Random-Oversampling (AUC = %0.2f)' % roc_auc_1)
plt.plot(fpr_0, tpr_0,  label='Random-Inicial Clases (AUC = %0.2f)' % roc_auc_o)
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```
