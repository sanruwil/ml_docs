# Read PDF

PyPDF2: Es una biblioteca de código abierto pura en Python que se puede utilizar para extraer texto y datos de archivos PDF. Proporciona una interfaz simple para dividir, fusionar, recortar y transformar los archivos PDF. Aquí hay un ejemplo de cómo leer el contenido de un archivo PDF:

```python
import PyPDF2

# Abre el archivo PDF
pdf_file = open('/Users/wilfredysantamariaruiz/Downloads/US_Curriculume-WilfredySantamaria-Actual.pdf', 'rb')

# Crea un objeto PDF
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Recorre todas las páginas del PDF
for page_num in range(len(pdf_reader.pages)):
    # Obtiene la página actual
    page = pdf_reader.pages[page_num]
    
    # Extrae el texto de la página
    text = page.extract_text()
    
    # Imprime el texto
    print(text)

# Cierra el archivo PDF
pdf_file.close()
```



pdfplumber: Es una biblioteca de Python que facilita la extracción de texto y datos de archivos PDF. Puede manejar archivos PDF complejos con tablas, formularios y otros elementos. Aquí hay un ejemplo de cómo leer el texto de un archivo PDF

```python
import pdfplumber

# Abre el archivo PDF
with pdfplumber.open('/Users/wilfredysantamariaruiz/Downloads/US_Curriculume-WilfredySantamaria-Actual.pdf') as pdf:
    # Recorre todas las páginas del PDF
    for page in pdf.pages:
        # Extrae el texto de la página
        text = page.extract_text()
        
        # Imprime el texto
        print(text)
```

```python
import fitz  # PyMuPDF

# Abre el archivo PDF
pdf_file = fitz.open('/Users/wilfredysantamariaruiz/Downloads/US_Curriculume-WilfredySantamaria-Actual.pdf')

# Recorre todas las páginas del PDF
for page_index in range(len(pdf_file)):
    # Selecciona la página actual
    page = pdf_file[page_index]
    
    # Extrae el texto de la página
    text = page.get_text()
    
    # Imprime el texto
    print(text)
    print("Pagina --> ",page_index)
```
