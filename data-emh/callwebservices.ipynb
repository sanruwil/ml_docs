```python
def main_webservices(json_data,url):
    requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
    #url = "https://localhost:8081/myService/pac008"  # Reemplaza con la URL del servicio REST que deseas consumir
    headers = {
        "Content-Type": "application/json",  # Especificamos que estamos enviando un JSON en el cuerpo
        "Authorization": "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpc3MiOiJFeGNlbHRpYSIsInN1YiI6IldlYlNlcnZpY2VzIiwiaWF0IjoxNTE2MjM5MDIyfQ.GFnvYa_GMuvoJl-R9No-F84zIKWKVdtUvth82bVoy-I"   # Ejemplo de encabezado de autorización con un token
    }
    try:
        # Realizamos la solicitud POST con el JSON y los encabezados
        response = requests.post(url, data=json.dumps(json_data), headers=headers,verify=False)
        #response = requests.get (url,headers=headers,verify=False)
        
        # Verificamos el código de respuesta HTTP para determinar si la solicitud fue exitosa
        if response.status_code == 200:
            print("Solicitud exitosa. Respuesta del servidor:")
            print(response.json())  # La respuesta del servidor en formato JSON
        else:
            print(f"Error - Código de estado: {response.status_code}")
            print(f"Error - Detalle: {response.text}")
    except requests.exceptions.RequestException as e:
        print(f"Error en la solicitud: {e}")

```
